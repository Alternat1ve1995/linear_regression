```python
X, y = df['km'], df['price']
```

```python
X = (X - X.min()) / (X.max() - X.min())
y = (y - y.min()) / (y.max() - y.min())
```

```python
plt.scatter(X, y);
```

```python
W0 = 0
W1 = 0
lr = 0.01
n = X.shape[0]
err = []
for _ in range(10000):
    pred = W0 + X * W1
    err += [sum((y - pred) ** 2) / (2 * n)]
    W0 -= (lr / n) * np.sum(pred - y)
    W1 -= (lr / n) * np.sum((pred - y) * X)
```

```python
plt.plot(list(range(len(err))), err);
```

```python
plt.scatter(X, y)
plt.plot(X, W0 + X * W1, color='red');
```
